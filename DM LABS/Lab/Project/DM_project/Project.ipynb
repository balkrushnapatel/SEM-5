# Import the pandas Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the dataset from this:....

```python
df = pd.read_csv("OnlineRetail.csv")
```

# Explore Dataset

```python
print("Shape:", df.shape)
print("\nColumn Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDescriptive Statistics:")
print(df.describe())
```

# Drop Missing Values

```python
df = df.dropna(subset=["CustomerID", "Description"])
```

# Drop Duplicate

```python
df = df.drop_duplicates()
```

# Remove Cancelled Orders (InvoiceNo starting with 'C')

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

# Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0]
```

# Convert InvoiceDate to Datetime Format

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

# Visualize & Handle Outliers in UnitPrice

```python
# View distribution using histogram
plt.hist(df['UnitPrice'], bins=100)
plt.title('Unit Price Distribution (Before Removing Outliers)')
plt.xlabel('UnitPrice')
plt.ylabel('Frequency')
plt.show()

# Remove UnitPrice outliers (e.g., > 100)
df = df[df['UnitPrice'] <= 100]

# Optional: Plot after cleaning
plt.hist(df['UnitPrice'], bins=50)
plt.title('Unit Price Distribution (After Removing Outliers)')
plt.xlabel('UnitPrice')
plt.ylabel('Frequency')
plt.show()
```

# Final Summary

```python
print("Final cleaned dataset shape:", df.shape)
print(df.describe())
df.head()
```
